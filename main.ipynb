```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df = pd.read_csv("archive/score_updated.csv")
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
x= df.drop(columns=["Scores"],axis=1)
y= df.drop(columns=["Hours"],axis=1)
```

```python
plt.figure(figsize=(14,8))
sns.set_style("dark")
sns.scatterplot(data=df,
                x="Hours",
                y="Scores",
                hue="Scores")
plt.xlabel("Hours students study")
plt.ylabel("Score done by student")
plt.title("Students study Hours time vs their outcomes")
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python

linear = LinearRegression()

linear.fit(X_train,y_train)
```

```python
print(f"The coefficent is {linear.coef_}")
print(f"The intercept is {linear.intercept_}")
```

```python
y_predict = linear.predict(X_train)
```

```python
plt.figure(figsize=(14,8))
sns.set_style("dark")
plt.scatter( X_train,y_train,color="red")
plt.plot(X_train,y_predict)
plt.xlabel("Hours students study")
plt.ylabel("Score done by student")
plt.title("Students study Hours time vs their outcomes")
```

```python
y_pre = linear.predict(X_test)
```

```python
plt.figure(figsize=(14,8))
sns.set_style("dark")
plt.scatter( X_test,y_test,color="red")
plt.plot(X_test,y_pre)
plt.xlabel("Hours students study")
plt.ylabel("Score done by student")
plt.title("Students study Hours time vs their outcomes")
```

```python
acc = linear.score(X_train,y_train)
mae = mean_absolute_error(y_test,y_pre)
mse = mean_squared_error(y_test,y_pre)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pre)
```

```python
print("Accuracy :" , round(acc * 100,2))
print("mean_squared_error :",round(mae,2))
print("MSE:      ", round(mse, 2))  
print("RMSE:     ", round(rmse, 2))  
print("R2 score: ", round(r2, 2))
```
